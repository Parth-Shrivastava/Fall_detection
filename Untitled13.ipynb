```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.patches as patches
import os
```

```python


train_img_path = '/content/drive/MyDrive/Fall_datset/images/train'
train_label_path = '/content/drive/MyDrive/Fall_datset/labels/train'


train_img_files = os.listdir(train_img_path)
train_img_files.sort()
r1='/content/drive/MyDrive/Fall_datset/images/train'
train_label_files = os.listdir(train_label_path)
train_label_files.sort()
r2= '/content/drive/MyDrive/Fall_datset/labels/train'
```

```python
# Display the corresponding image
import matplotlib.pyplot as plt

img_path = os.path.join(r1, train_img_files[24])
img = plt.imread(img_path)

plt.imshow(img)
plt.axis('off')
plt.title(f"Image: {train_img_files[24]}")
plt.show()
```

```python
with open(os.path.join(train_label_path, train_label_files[24]), 'r') as file:
    r = file.readlines()
print(r)
bounding_boxes = []
for j in r:
    j = j.split()
    bounding_boxes.append([int(j[0]), float(j[1]), float(j[2]), float(j[3]), float(j[4])])
```

```python
import matplotlib.patches as patches

fig, ax = plt.subplots(1)
ax.imshow(img)


for normalized_bbox in bounding_boxes:
    image_height, image_width, _ = img.shape
    xmin, ymin, width, height = normalized_bbox[1:]

    xmin = int(xmin * image_width)
    ymin = int(ymin * image_height)
    width = int(width * image_width)
    height = int(height * image_height)

    xmin = max(0, xmin)
    ymin = max(0, ymin)

    color = 'r' if normalized_bbox[0] == 0 else 'b' if normalized_bbox[0] == 1 else 'g'
    rect = patches.Rectangle((xmin - width/2, ymin - height/2), width, height,
                             linewidth=2, edgecolor=color, facecolor='none')
    ax.add_patch(rect)

plt.title(f"Image with Bounding Box: {train_img_files[24]}")
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
df = pd.DataFrame()
list_of_paths = []
list_of_classes = []
for train_file in train_img_files:
    list_of_paths.append(os.path.join(train_img_path, train_file))
    list_of_classes.append(1 if train_file.startswith('f') else 0)

df["Path"] = list_of_paths
df["Class"] = list_of_classes


list_of_images = []
target_size = (150, 150)
for path in df["Path"]:
    image = Image.open(path).resize(target_size)
    image_array = np.array(image)
    list_of_images.append(image_array)

df['Image_arr'] = list_of_images
print(df['Image_arr'][0].shape)
```

```python
df['Image_arr'][0].shape
```

```python
from tensorflow import keras

model = keras.Sequential([

    keras.layers.Conv2D(16, (3, 3), padding='same', input_shape=(150, 150, 3), activation='relu'),
    keras.layers.MaxPooling2D(),


    keras.layers.Conv2D(32, (3, 3), padding='same', activation='relu'),
    keras.layers.MaxPooling2D(),


    keras.layers.Conv2D(64, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D(),


    keras.layers.Flatten(),

    keras.layers.Dense(128, activation='relu'),

    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
X_train = np.array(df['Image_arr'].tolist())
y_train = np.array(df['Class'])
model.fit(X_train, y_train, epochs=20)
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models

base_model = VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

for layer in base_model.layers:
    layer.trainable = False

model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))  # Binary classification, so use 1 output with sigmoid activation
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python

def predict_image(img_path, model, target_size=(150, 150)):
    """
    Predicts whether a given image is a fall or not fall.
    Args:
        img_path (str): Path to the image.
        model (keras.Model): Trained model.
        target_size (tuple): Resize dimensions.
    """
    image = Image.open(img_path).resize(target_size)
    image_array = np.array(image)

    if image_array.shape[-1] != 3:
        print("Skipping grayscale image or corrupted:", img_path)
        return

    image_array = image_array / 255.0
    image_array = image_array.reshape(1, *target_size, 3)

    pred = model.predict(image_array)[0][0]
    label = "FALL" if pred > 0.5 else "NO FALL"
    print(f"Prediction: {label} (Confidence: {pred:.4f})")

    plt.imshow(image)
    plt.title(f"Prediction: {label}")
    plt.axis('off')
    plt.show()
```
